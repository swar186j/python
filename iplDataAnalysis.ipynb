```python
#returns top 5 teams in the key format
top_winners = ipl['winner'].value_counts()[0:5].keys()
print(top_winners)

top5=ipl['winner'].value_counts()[0:5]
print(top5)
```

```python
#plot a bar graph for winners vs their count
plt.figure(figsize=(8,5))

plt.bar(list(top_winners),list(top5), color='g')
```
